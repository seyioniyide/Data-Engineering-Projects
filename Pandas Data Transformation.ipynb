```python
#Pivot Tables
data = {
    'A': ['foo', 'foo', 'foo', 'bar', 'bar', 'bar'],
    'B': ['one', 'one', 'two', 'two', 'one','one' ],
    'C': ['x', 'y', 'x', 'y', 'x', 'y'],
    'D': [1, 3, 2, 5, 4, 1]
}
df = pd.DataFrame(data)
df
```

```python
import numpy as np
pivot_df = df.pivot_table(values ='D', #to aggregate the values of which column
                         index='A',    # column to use as index
                         columns=['C'], #value of which column as the new column
                         aggfunc=np.sum)  # aggregation function to use
pivot_df
```

```python
# resetting the index
pivot_df.reset_index()
```

```python
#performing pivot table action on our weather dataset
# what is the mean temperature recored per month using pivot table
mean_temperature_df = weather_df.pivot_table(values='Temp (C)', index=weather_df['Date/Time'].dt.month, aggfunc=np.mean)
mean_temperature_df
```

```python
#Group BY
mean_temperature_df2 = weather_df.groupby(weather_df['Date/Time'].dt.month).agg(np.mean).reset_index()
mean_temperature_df2
```
